```python
import numpy as np 
import matplotlib.pyplot as plt 
  
  
Turma_ = ['A', 'B'] 
  
data = [26, 24] 
  
  
explode = (0.1, 0.0) 
colors = ( "orange", "cyan") 
wp = { 'linewidth' : 1, 'edgecolor' : "green" } 
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%\n({:d} g)".format(pct, absolute) 
fig, ax = plt.subplots(figsize =(10, 7)) 
wedges, texts, autotexts = ax.pie(data,  
                                  autopct = lambda pct: func(pct, data), 
                                  explode = explode,  
                                  labels = Turma_, 
                                  shadow = True, 
                                  colors = colors, 
                                  startangle = 90, 
                                  wedgeprops = wp, 
                                  textprops = dict(color ="magenta")) 
ax.legend(wedges, Turma_, 
          title ="Turma_", 
          loc ="center left", 
          bbox_to_anchor =(1, 0, 0.5, 1)) 
  
plt.setp(autotexts, size = 8, weight ="bold") 
ax.set_title("Porcentagem das Turmas A e B") 
plt.show() 
```

# **Análise de duas variáveis quantitativas e grau de relação linear entre elas (quantificar essa relação)**

```python
Newdf = dataframe.drop(columns=['Id', 'Turma', 'Sexo', 'Idade', 'Alt', 'Filhos', 'Fuma', 'Tolernacia','Cine','OpCine', 'TV', 'OpTv'])
Newdf
```

```python
import statistics
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from   sklearn.linear_model import LinearRegression
from   sklearn.metrics import r2_score
import statsmodels.api as sm
import os
```

```python
plt.figure(figsize = (16,8))
plt.scatter(
    Newdf['Peso'], 
    Newdf['Exer'], 
    c='red')
plt.xlabel("  Peso em quilogramas por pessoa")
plt.ylabel(" Hora de  prática de exercício semanal por pessoa")

plt.show()

```

```python
X = dataframe['Peso'].values.reshape(-1,1)
y = dataframe['Exer'].values.reshape(-1,1)


reg = LinearRegression()
reg.fit(X, y)


print("O modelo é: Peso x Exer = {:.5} + {:.5}X".format(reg.intercept_[0], reg.coef_[0][0]))

```

```python
f_previsaoes = reg.predict(X)


plt.figure(figsize = (16,8))
plt.scatter(
    dataframe['Peso'], 
    dataframe['Exer'], 
    c='red')


plt.plot(
    Newdf['Peso'],
    f_previsaoes,
    c='blue',
    linewidth=3,
    linestyle=':'
)

plt.xlabel("Peso em quilogramas por pessoa")
plt.ylabel("  Hora de  prática de exercício semanal por pessoa")
plt.show()
```
